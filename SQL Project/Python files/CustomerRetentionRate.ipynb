```python
#retention comparison of customers in the first half of 2017 to second half of 2017
#convert string to datetime and arrange two dataframes based on date range
s1H12017 = pd.to_datetime("2017-01-01").date()
e1H12017 = pd.to_datetime("2017-06-30").date()
df1=newdf[(newdf['Date'] > s1H12017) & (newdf['Date'] < e1H12017)].reset_index()
s2H2017 = pd.to_datetime("2017-07-01").date()
e2H2017 = pd.to_datetime("2017-12-31").date()
df2=newdf[(newdf['Date'] > s2H2017) & (newdf['Date'] < e2H2017)].reset_index() 


#drop duplicates based on customer's unique ID
df1Dropped=df1.drop_duplicates("customer_unique_id")
df2Dropped=df2.drop_duplicates("customer_unique_id")

#move unique IDs to list to be compared and then delete the index column that appeared after resetting indexes
l1=df1Dropped['customer_unique_id'].tolist()
l2=df2Dropped['customer_unique_id'].tolist()
del df1Dropped["index"]
del df2Dropped["index"]


display(df1Dropped,df2Dropped)
```

```python
l3=list(set(l2).intersection(set(l1))) 
len(l3)
```

```python
def retention(olist, nlist):
    retlist=list(set(nlist).intersection(set(olist))) 
    return(retlist)
```

```python
len(retention(l1,l2))
```

```python
def retDeci(olist, nlist):
    retlist=list(set(nlist).intersection(set(olist))) 
    return((len(retlist)/len(olist)))
```

```python
display(retDeci(l1,l2))
```

```python
tb1 = pd.DataFrame()
nV=["Jan:Jun 2017 - July:Dec 2017",
    "July:Dec 2017- Jan:Jun 2018",
    "Jan:Jun 2018 - July:Dec 2018",
    "Jan:Jun 2017 - July:Dec 2018",
    "All of 2017-All of 2018"]
vV=[len(retention(l1,l2)),
    len(retention(l8,l9)),
        len(retention(l4,l5)),
        len(retention(l1,l5)),
        len(retention(l6,l7))]
dV=[retDeci(l1,l2),
    retDeci(l8,l9),
        retDeci(l4,l5),
        retDeci(l1,l5),
        retDeci(l6,l7)]
tb1["Retention Range"]=nV
tb1["# of People Retained Over"]=vV
tb1["Retention Rate"]=dV

display(tb1)
```
