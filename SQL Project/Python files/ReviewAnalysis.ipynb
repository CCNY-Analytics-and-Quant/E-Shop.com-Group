```python
import pandas as pd
import pip
pip.main(["install","openpyxl"])
import openpyxl as xls
from datetime import datetime

reviewInfo=pd.read_csv(r"C:\Users\osamw\Documents\Python for Buiness Analytics\review data.csv")
orderBase=pd.read_csv(r"C:\Users\osamw\Documents\Python for Buiness Analytics\order base data.csv")
customerData=pd.read_csv(r"C:\Users\osamw\Documents\Python for Buiness Analytics\customer base data.csv")
productInfo=pd.read_csv(r"C:\Users\osamw\Documents\Python for Buiness Analytics\product info data.csv")
productInEng=pd.read_csv(r"C:\Users\osamw\Documents\Python for Buiness Analytics\product name translator data.csv")
orderItems=pd.read_csv(r"C:\Users\osamw\Documents\Python for Buiness Analytics\order items data.csv")

display(reviewInfo,orderBase,customerData,productInfo,productInEng,orderItems)
```

```python
orderMerge=orderBase.drop(columns=['date_approved','date_shipped','date_arrived','date_estimated'])
#orderMerge2=orderMerge.loc[orderMerge["order_Status"
customerMerge=customerData.drop(columns=['customer_unique_id','customer_zip_code_prefix','customer_city'])
productMerge=productInfo.drop(columns=['product_photos_qty'])
orItemMerge=orderItems.drop(columns=['seller_id','shipping_limit_date','price'])


tb1=pd.merge(reviewInfo,orderMerge,on="order_id",how="left")
tb2=pd.merge(tb1,customerMerge,on="customer_id",how="inner")
tb3=pd.merge(tb2,orItemMerge,on="order_id", how="left",indicator=True)

display(tb1,tb2,tb3)
```

```python
tb3.sort_values(by=['order_item_id'], ascending=False)
```

```python
t1=pd.merge(reviewInfo[reviewInfo.order_id.duplicated(keep=False)],orderItems, on="order_id", how="left",indicator=True)
 
display(reviewInfo[reviewInfo.order_id.duplicated(keep=False)],t1.sort_values(by=['order_item_id'], ascending=False))
```

```python
tb3v2=tb3.drop_duplicates("order_id",keep="first").reset_index()
del tb3v2["index"]
display(tb3v2)
```

```python
tb3v3=pd.merge(tb3v2,productMerge, on="product_id", how="left")

tb3v3
```

```python
tb3v3.columns = ['review_id',
                 'order_id',
                 'date_sent',
                 'review_score',
                 'customer_id',
                 'order_status',
                 'customer_state',
                 'order_item_id',
                 'product_id',
                 '_merge',
                 'product_name']



tb3vF=pd.merge(tb3v3,productInEng, on="product_name", how="left")

display (tb3vF)
```

```python
excelFile= pd.ExcelWriter('Review Analysis Both Years Combined.xlsx')

tb3vF.to_excel(excelFile)

excelFile.save()

tb3vF.to_csv('Review Analysis Both Years Combined.csv')
```

```python
tb3vF['Date'] = pd.to_datetime(tb3vF['date_sent']).dt.date

s2017 = pd.to_datetime("2017-01-01").date()
e2017 = pd.to_datetime("2017-12-31").date()
df1=tb3vF[(tb3vF['Date'] > s2017) & (tb3vF['Date'] < e2017)].reset_index()
s2018 = pd.to_datetime("2018-01-01").date()
e2018 = pd.to_datetime("2018-12-31").date()
df2=tb3vF[(tb3vF['Date'] > s2018) & (tb3vF['Date'] < e2018)].reset_index()

del df1["index"]
del df2["index"]

display(df1,df2)
```
