```python
import pandas as pd
from datetime import datetime
import pip
pip.main(["install","openpyxl"])
import openpyxl as xls



orderBase=pd.read_csv(r"C:\Users\osamw\Documents\Python for Buiness Analytics\order base data.csv")
display(orderBase)
```

```python
#place holder for later
days_ahead_or_behind=[ ]

#orderBWI= orderBase without invoices (delivered items only) 
orderBWI=orderBase.loc[orderBase['order_status']=='delivered']

#loop that caluculates difference of days from date_arrived to date_estimated
for num in range (orderBWI.shape[0]):
    x=datetime.strptime(orderBWI.iloc[num,6], "%Y-%m-%d  %H:%M:%S.0000000")-datetime.strptime(orderBWI.iloc[num,5], "%Y-%m-%d  %H:%M:%S.0000000")
    days_ahead_or_behind+= [x.days + 1]
```

```python
orderID=orderBWI['order_id'].tolist()#converted order_id values into a list 
dA = orderBWI['date_arrived'].tolist()
dE= orderBWI['date_estimated'].tolist()

ab = pd.DataFrame(orderID, columns={'Order ID'}) #create new dataframe

ab['Day Arrived']=dA #add list to dataframe
ab['Arrival Estimated']=dE
ab['Days Ahead or Behind Schedule']=days_ahead_or_behind 

display(ab)
```

```python

excelFile= pd.ExcelWriter('Days ahead or behind.xlsx')

ab.to_excel(excelFile)

excelFile.save()

ab.to_csv('Days ahead or behind.csv')
```
