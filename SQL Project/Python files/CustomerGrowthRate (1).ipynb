```python
import pandas as pd
import pip
pip.main(["install","openpyxl"])
import openpyxl as xls
from datetime import datetime

orderBase=pd.read_csv(r"C:\Users\osamw\Documents\Python for Buiness Analytics\order base data.csv")
customerData=pd.read_csv(r"C:\Users\osamw\Documents\Python for Buiness Analytics\customer base data.csv")
orderItems=pd.read_csv(r"C:\Users\osamw\Documents\Python for Buiness Analytics\order items data.csv")

display(orderBase,customerData,orderItems)
```

```python
oBaseMerge=orderBase.drop(columns=['date_shipped','date_arrived','date_estimated'])
custMerge=customerData.drop(columns=['customer_zip_code_prefix','customer_city'])
oItems=orderItems.drop(columns=['seller_id','shipping_limit_date','price'])

tb1=pd.merge(custMerge,oBaseMerge, on="customer_id", how="left")

tb1
```

```python
tb2=pd.merge(tb1,oItems, on="order_id", how="left")
tb2
```

```python
tb3['Date'] = pd.to_datetime(tb3['date_approved']).dt.date
tb3['Quarter'] = pd.to_datetime(tb3['Date']).dt.quarter
tb3['Quarter_Info'] = pd.to_datetime(tb3['Date']).dt.to_period('Q')
tb3
```

```python
tb4=tb3.drop(columns=['Quarter','date_approved'])
tb4
```

```python
y1q1=tb4.loc[tb4['Quarter_Info']=='2017Q1'].reset_index()
del y1q1['index']
y1q1

y1q2=tb4.loc[tb4['Quarter_Info']=='2017Q2'].reset_index()
del y1q2['index']


y1q3=tb4.loc[tb4['Quarter_Info']=='2017Q3'].reset_index()
del y1q3['index']


y1q4=tb4.loc[tb4['Quarter_Info']=='2017Q4'].reset_index()
del y1q4['index']


y2q1=tb4.loc[tb4['Quarter_Info']=='2018Q1'].reset_index()
del y2q1['index']


y2q2=tb4.loc[tb4['Quarter_Info']=='2018Q2'].reset_index()
del y2q2['index']


y2q3=tb4.loc[tb4['Quarter_Info']=='2018Q3'].reset_index()
del y2q3['index']

y1=tb4.loc[(tb4['Quarter_Info']>='2017Q1')&(tb4['Quarter_Info']<='2017Q4')].reset_index()
del y1['index']

y2=tb4.loc[(tb4['Quarter_Info']>='2018Q1')&(tb4['Quarter_Info']<='2018Q3')].reset_index()
del y2['index']

#y2q4=tb4.loc[tb4['Quarter_Info']=='2018Q4'].reset_index() ##NO VALUES FOR THE 4TH QUARTER OF 2018##
#del y2q4['index']

display(y1q1,y1q2,y1q3,y1q4,y2q1,y2q2,y2q3,y1.sort_values(by=['Quarter_Info'], ascending=False),y2.sort_values(by=['Quarter_Info'], ascending=False))
```

```python
def pctC (olddf,newdf):
    change= ((newdf.shape[0]-olddf.shape[0])/olddf.shape[0])
    return(change)
```

```python
newcol=[pctC(y1q1,y1q2),pctC(y1q2,y1q3),pctC(y1q3,y1q4),pctC(y1q4,y2q1),pctC(y2q1,y2q2),pctC(y2q2,y2q3),pctC(y1,y2)]
```

```python
names=['2017Q1-2017Q2','2017Q2-2017Q3','2017Q3-2017Q4','2017Q4-2018Q1','2018Q1-2018Q2','2018Q2-2018Q3','End of 2017-Q3 of2018']

ab = pd.DataFrame(names,columns={'Year Range'})
ab['Growth Rate']=newcol
ab
```

```python
excelFile= pd.ExcelWriter('Customer Growth Rate General.xlsx')

ab.to_excel(excelFile)

excelFile.save()

ab.to_csv('Customer Growth Rate General.csv')
```
